```python
import pandas as pd
import numpy as np
from itertools import permutations
from mlxtend.preprocessing import TransactionEncoder
from apyori import apriori


data=pd.read_csv("/content/drive/MyDrive/THESIS/Report.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
data['CLASS_NAME'].isna().sum()
```

```python

df=data.drop(["check","QTY","RATE","TOTAL","COST_PRICE"],axis=1)
df.head()
```

```python
transactions=df['CLASS_NAME']
```

```python
transactions=data['CLASS_NAME'].apply (lambda t: t.split(','))
```

```python
Transactions=list(transactions)
print(Transactions[0])
print(type(Transactions))
```

```python
print(len(Transactions))
```

```python
print(type(Transactions))
```

```python
encoder=TransactionEncoder().fit(Transactions)
```

```python
onehot=encoder.transform(Transactions)
```

```python
onehot=pd.DataFrame(onehot,columns=encoder.columns_)
print(onehot)
```

```python
print(onehot.mean())
```

Apriori algo

```python
result=apriori(Transactions,min_support=.0500, min_confidence=.2,min_lift=1.002,min_length=2)
results=list(result)
%time
```

```python
print("There are {} Relation derived.".format(len(results)))
```

```python
for i in range (0,len(results)):
 print(results[i][0])
```

```python
result=apriori(Transactions,min_support=.0300, min_confidence=.250,min_lift=1.002,min_length=2)
results=list(result)
%time
```

```python
print("There are {} Relation derived.".format(len(results)))
```

```python
for i in range (0,len(results)):
 print(results[i][0])
```

FP growth algo


```python
from mlxtend.frequent_patterns import fpgrowth

res=fpgrowth(onehot,min_support=0.05, use_colnames=True)

res.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules


res=association_rules(res, metric="lift", min_threshold=1)


res
%time
```

```python
res.sort_values("confidence",ascending=False)
```

```python
from mlxtend.frequent_patterns import fpgrowth

res=fpgrowth(onehot,min_support=0.03, use_colnames=True)

res.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules


res=association_rules(res, metric="lift", min_threshold=1)


res
%time
```

```python
res.sort_values("confidence",ascending=False)
```

```python
from csv import reader
list = []

with open("/content/drive/MyDrive/THESIS/prefixspan.csv", 'r') as read_obj:
    
    csv_reader = reader(read_obj)
    customerlist=[]
    sequence = []
    anotherList=[]
    for row in csv_reader:

        list2 = []

        for x in row:

              if x.isdigit():
                print("before")
                print(x)
                if  len(customerlist) == 0 or  customerlist[len(customerlist) - 1] != x:

                     if len(customerlist) != 0:
                            list.append(Row(sequence=anotherList))
                     customerlist.append(x)
                     sequence=[]
                     anotherList=[]

              if x != "" and not x.isdigit() :

                  list2.append(x)

        anotherList.append(list2)
list.append(Row(sequence=anotherList))
print(list)



df = sc.parallelize(list).toDF()

prefixSpan = PrefixSpan(minSupport=0.05, maxPatternLength=2,
                        maxLocalProjDBSize=32000000)


prefixSpan.findFrequentSequentialPatterns(df).sort(desc("freq")).show(50,False)
```
